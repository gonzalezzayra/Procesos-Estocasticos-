**Networkx**

NetworkX es una paquetería de Python para la creación, manipulación y estudio de la estructura, dinámica y funciones de redes complejas

Se define la librería de "networkx" y "matplotlib.pyplot" para los diagramas de los grafos.

````````python
import networkx as nx
import matplotlib.pyplot as plt
````````

````````python
#Para definir el grafo dirigido
G = nx.DiGraph()
````````

Estos es para añadir los nodos y sus arcos, así como su 'capacity', aquí se define de donde a donde va el arco.

````````python
G.add_edge("A", "B", capacity = 2) #A -> B
G.add_edge("B", "C", capacity = 2) #B -> C
G.add_edge("C", "A", capacity = 2) #C -> A
````````

````````python
#Para dibujar el grafo y realizar la gráfica/grafo
nx.draw_networkx(G)
plt.title("Visualización del Grafo")
plt.show()
````````

**Otra forma de hacerlo**

````````python
#Agregar los nodos
G.add_nodes_from(["A", "B", "C"])
````````

````````python
#Agregar los arcos y se pone de donde a donde va los arcos
G.add_edges_from([("A", "B"), ("B", "C"), ("C", "A")])
````````

````````python
#Aquí para que quedé de una forma fija, se llama disposición circular
diagrama = nx.circular_layout(G)
nx.draw_networkx(G, diagrama, with_labels=True)
plt.show()
````````

A diferencia del anterior, se modifica las características del grafo y que tan diferente se ve

````````python
diagrama = nx.circular_layout(G)
nx.draw_networkx(G, diagrama, with_labels=True, node_color="lightblue", node_size=2000, arrows=True)
plt.title("Visualización del Grafo")
plt.show()
````````

Este comando da el conjunto de nodos

````````python
nx.nodes(G)
````````

Este comando da el conjunto de arcos

````````python
nx.edges(G)
````````

Este comando da el número de nodos

````````python
nx.number_of_nodes(G)
````````

Este comando da el número de arcos

````````python
nx.number_of_edges(G)
````````

Este comando da el número de arcos a los que esta conectado, en otros terminos el número de aristas de cada nodo

````````python
nx.degree(G)
````````
