**Caminata aleatoria**

Instrucción: Implementar  una caminata aleatoria 1D, con probabilidad p de moverse a la derecha y probabilidad $q$ de moverse a la izquierda, tal que

$$p + q = 1$$

Se inicia con importar la librería *random*, para generar numeros aleatorios

````python
import random as r
````

*Pb* es para la probabilidad, ya que con random se genera un numero entre 0 y 1, ya que la probabilidad no puede ser mayor a 100 o 1, esto no funciona; es por ello que *Pb* funciona como la probabilidad.

*Pb* va ir variando, no es un valor fijo.

````python
pb=r.random()
print(pb)
````

Se define la función de *caminata*

*u* es igual o se le asigna el valor del número aleatorio/random, con un ciclo si, veremos si el valor de u es mayor a la probabilidad, te devuelve un *1* y sino es así se devuelve *-1*.

````python
def caminata(pb):
  u=r.random()
  if u<pb:
    return 1
  else:
    return -1
````

````python
caminata(pb) #Prueba
````

Se define un bucle

````python
pasos=[] #Lista vacía para guardar los valores del ciclo for
sum=0 #Se inicializa la suma o contador
n=100 #numero de pasos
l=[0] #Esto funciona para que empiece con el 0

#Ciclo for
for i in range(n):
  pasos.append(caminata(pb)) #en la lista de pasos se irá guardando los valores de la función "caminata" en el valor de 'pb'
  sum+=caminata(pb) #se va sumando los valores, que puede ser 1 y -1
  l.append(sum)
````

**Primera pregunta**

````python
def caminata3(pb):
    pasos = []
    sum = 0
    l = [0]
    ll= []
    n = 100

    for i in range(n):
        u = r.random()
        if u < pb:
            paso = 1
        else:
            paso = -1

        pasos.append(paso)
        sum += paso
        l.append(sum)
        ll.append(abs(sum))

    punto_l = max(ll)
    return punto_l
````

````python
#El punto más lejano

print("El punto más lejano es:", caminata3(pb))
````

**Segunda pregunta**

Se modifica el código anterior.

Se importa la libreria de *mean*, para después sacar el promedio

````python
from statistics import mean #Esto es para obtener el promedio
````

````python
from statistics import mean

def caminata3prom(pb):
    pasos = []
    sum = 0
    l = [0]
    ll = []
    n = 100

    for i in range(n):
        u = r.random()
        if u < pb:
            paso = 1
        else:
            paso = -1

        pasos.append(paso)
        sum += paso
        l.append(sum)
        ll.append(abs(sum))

    promedio = mean(ll)
    return promedio
````

````python
print("Distancia promedio:", caminata3prom(pb))
````

**Tercera pregunta**

````python
def caminata3or(pb):
    pasos = []
    sum = 0
    l = [0]
    n = 100

    for i in range(n):
        u = r.random()
        if u < pb:
            paso = 1
        else:
            paso = -1

        sum += paso
        l.append(sum)

    ceros = l.count(0) #Se cuenta cuanto ceros hay en la lista

    if ceros > 1:
        print("Regresa al origen")
    else:
        print("No regresaal origen")
````
