**Caminata aleatoria**

Instrucción: Implementar  una caminata aleatoria 1D, con probabilidad p de moverse a la derecha y probabilidad $q$ de moverse a la izquierda, tal que

$$p + q = 1$$

Se inicia con importar la librería *random*, para generar numeros aleatorios

````python
import random as r
````

*Pb* es para la probabilidad, ya que con random se genera un numero entre 0 y 1, ya que la probabilidad no puede ser mayor a 100 o 1, esto no funciona; es por ello que *Pb* funciona como la probabilidad.

*Pb* va ir variando, no es un valor fijo.

````python
pb=r.random()
print(pb)
````

Se define la función de *caminata*

*u* es igual o se le asigna el valor del número aleatorio/random, con un ciclo si, veremos si el valor de u es mayor a la probabilidad, te devuelve un *1* y sino es así se devuelve *-1*.

````python
def caminata(pb):
  u=r.random()
  if u<pb:
    return 1
  else:
    return -1
````

````python
caminata(pb) #Prueba
````

Se define un bucle

````python
pasos=[] #Lista vacía para guardar los valores del ciclo for
sum=0 #Se inicializa la suma o contador
n=100 #numero de pasos
l=[0] #Esto funciona para que empiece con el 0

#Ciclo for
for i in range(n):
  pasos.append(caminata(pb)) #en la lista de pasos se irá guardando los valores de la función "caminata" en el valor de 'pb'
  sum+=caminata(pb) #se va sumando los valores, que puede ser 1 y -1
  l.append(sum)
````

Ahora se definirá una segunda caminata, pero ahora cambia el tamaño/numero de pasos.

````python
def caminata2(pb):
  pasos=[]
  sum=0
  n=10**3
  l=[0]
  for i in range(n):
    pasos.append(caminata(pb))
    sum+=caminata(pb)
    l.append(sum)
  plt.plot(l, color="pink")
  plt.xlabel("pasos")
  plt.ylabel("distancia")
  plt.show()
caminata2(pb)
````
